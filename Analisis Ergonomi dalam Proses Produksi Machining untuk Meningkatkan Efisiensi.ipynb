# Analisis Ergonomi dalam Proses Produksi Machining untuk Meningkatkan Efisiensi

**Pendahuluan**


Proses produksi machining merupakan bagian integral dalam industri manufaktur yang menghasilkan berbagai komponen dan produk. Mesin CNC (Computer Numerical Control) telah mengubah cara produksi machining berlangsung, meningkatkan akurasi dan kecepatan. Meskipun demikian, untuk mencapai efisiensi maksimum dalam produksi, perhatian yang serius terhadap faktor-faktor ergonomi adalah suatu keharusan. Dalam makalah ini, kita akan membahas pentingnya analisis ergonomi dalam proses produksi machining dan bagaimana faktor-faktor ini dapat ditingkatkan untuk meningkatkan efisiensi.



**Analisis Ergonomi**


Waktu Operasi: Data waktu operasi mesin adalah parameter kunci dalam produksi machining. Analisis ergonomi harus mempertimbangkan apakah waktu operasi yang lama dapat meningkatkan risiko kelelahan operator. Perubahan dalam desain alat dan pemilihan alat yang tepat dapat membantu mengurangi waktu operasi.

Kelelahan Operator: Kelelahan operator adalah faktor penting dalam analisis ergonomi. Nilai tingkat kelelahan yang tinggi dapat mengakibatkan kesalahan manusia dan penurunan produktivitas. Melalui pelatihan dan perubahan desain alat, tingkat kelelahan operator dapat dikelola dengan lebih baik.

Pencahayaan dan Suhu: Pencahayaan yang baik dan suhu yang nyaman di tempat kerja adalah aspek penting dalam kenyamanan dan produktivitas operator. Pengukuran pencahayaan dan suhu serta penyesuaian yang tepat dapat membantu menciptakan lingkungan yang lebih baik untuk pekerja.

Ketinggian Meja: Ketinggian meja mesin juga merupakan faktor ergonomi yang perlu diperhatikan. Meja yang disesuaikan dengan benar dengan tinggi operator dapat mengurangi ketegangan fisik dan meningkatkan produktivitas.

Kualitas Produk: Kualitas produk yang dihasilkan dalam proses machining adalah hasil langsung dari faktor-faktor ergonomi dan operasional. Kualitas dapat ditingkatkan dengan mengidentifikasi dan mengatasi masalah ergonomi yang mungkin memengaruhi hasil produksi.

**Meningkatkan Efisiensi**


Dalam analisis ergonomi, tujuan utama adalah meningkatkan efisiensi produksi. Dengan memperhatikan faktor-faktor ergonomi seperti yang dijelaskan di atas, langkah-langkah berikut dapat diambil untuk meningkatkan efisiensi:

Optimasi Desain Alat: Memilih atau merancang alat yang lebih efisien dan ergonomis untuk penggunaan operator.

Pelatihan Operator: Memberikan pelatihan kepada operator untuk meningkatkan keterampilan dan kesadaran ergonomi.

Pengawasan Faktor Lingkungan: Memastikan kondisi lingkungan kerja yang nyaman, termasuk pencahayaan yang baik dan suhu yang sesuai.

Pengukuran dan Monitoring Reguler: Melakukan pengukuran rutin terhadap faktor-faktor ergonomi dan melacak perubahan dalam waktu.

Kontrol Kualitas: Menggunakan pengamatan ergonomi untuk memperbaiki kualitas produk dan mengurangi limbah.

# Model Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
labels = dat['Kualitas Produk (skala 1-5)']

dat = dat.drop(columns="Kualitas Produk (skala 1-5)")
```

```python
X=dat
y=labels

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                
                                                    test_size = 0.2,
                                                    random_state = 0)


print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Melakukan scalling (Standarisasi) pada feature

sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_train2.columns = X_train.columns.values
X_train2.index = X_train.index.values
X_train = X_train2

X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_test2.columns = X_test.columns.values
X_test2.index = X_test.index.values
X_test = X_test2
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score



# Membangun model regresi linear
model = LinearRegression()

# Melatih model pada data latih
model.fit(X_train, y_train)

# Menguji model pada data uji
y_pred = model.predict(X_test)

# Mengukur kinerja model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Hasil pengujian model:{y_pred}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

hasil pengujian model Anda menunjukkan Mean Squared Error (MSE) yang sangat rendah (mendekati nol) dan R-squared (R2) yang mendekati 1.0 , menunjukan bahwa hasil model sangat baik.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Membangun model Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
logistic_predictions = logistic_model.predict(X_test)

# Mengukur kinerja model Logistic Regression
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
logistic_classification_report = classification_report(y_test, logistic_predictions)

# Membangun model Decision Tree
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
decision_tree_predictions = decision_tree_model.predict(X_test)

# Mengukur kinerja model Decision Tree
decision_tree_accuracy = accuracy_score(y_test, decision_tree_predictions)
decision_tree_classification_report = classification_report(y_test, decision_tree_predictions)

# Membangun model Random Forest
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)
random_forest_predictions = random_forest_model.predict(X_test)

# Mengukur kinerja model Random Forest
random_forest_accuracy = accuracy_score(y_test, random_forest_predictions)
random_forest_classification_report = classification_report(y_test, random_forest_predictions)

# Menampilkan hasil pengujian untuk ketiga model
print("Hasil pengujian model Logistic Regression:")
print(f"Accuracy: {logistic_accuracy}")
print("Classification Report:")
print(logistic_classification_report)

print("\nHasil pengujian model Decision Tree:")
print(f"Accuracy: {decision_tree_accuracy}")
print("Classification Report:")
print(decision_tree_classification_report)

print("\nHasil pengujian model Random Forest:")
print(f"Accuracy: {random_forest_accuracy}")
print("Classification Report:")
print(random_forest_classification_report)
```
